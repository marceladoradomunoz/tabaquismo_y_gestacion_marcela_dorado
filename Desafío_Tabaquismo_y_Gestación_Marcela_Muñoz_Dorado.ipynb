```python
#Creación de función que le otorga a X las la media, varianza, el mínimo y el máximo.

def get_psd(x):
  mean=x.mean()
  sigma=x.std()
  minimo=x.min()
  maximo=x.max()
  x = np.linspace(minimo, maximo)
  pdf = norm.pdf(x, loc=mean, scale=sigma)
  return(x,pdf)
```

```python
#Gráfico de comparación entre embarazadas fumadoras y no fumadoras

x,pdf=get_psd(bebes_mother_smokers)
plt.plot(x, pdf, color='red', label='Madres NO Fumadoras')
x,pdf=get_psd(bebes_mother_no_smokers)
plt.plot(x, pdf, color='blue', label='Madres Fumadoras')
plt.hist(bebes_mother_smokers, bins=20, density=True, alpha=0.2, color='red')
plt.hist(bebes_mother_no_smokers, bins=20, density=True, alpha=0.2, color='blue')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad normalizada')
plt.title('Comparación de pesos de fetos de embarazadas fumadoras y no fumadoras')
plt.legend()
plt.show()
```

####**C) UTILIZANDO ESTE MODELO, SIMULA 1000 MUESTRAS DE TAMAÑO 8, Y CALCULA SUS MEDIAS MUESTRALES.**
####**D) UTILIZANDO LOS DATOS DIRECTAMENTE, EXTRAE 1000 MUESTRAS DE TAMAÑO 8 Y CALCULA SUS RESPECTIVAS MEDIAS MUESTRALES**.

```python
#PREGUNTA 5 LETRA A): PROBABILIDAD DE QUE UNA MADRE ESCOGIDA AL AZAR SE FUMADORA.

smoking_mothers = df_baby['Maternal.Smoker'].dropna()
smoking = np.where(smoking_mothers, 1, 0)
prob = sum(smoking) / len(smoking)
print(f'La probabilidad de que una madre escogida al azar sea fumadora es de: {prob}%.-')
```

```python
#PREGUNTA 5 LETRA B): UTILIZANDO FUNCIÓN SAMPLE 5 Y GRAFICANDO

#1ª PARTE: código filtrando y utilizando función sample.
smoking_mothers = df_baby['Maternal.Smoker'].sample(5)
smoking_mothers = smoking_mothers.astype(int)
print(smoking_mothers)

plt.hist(smoking_mothers, alpha=0.5, color='pink')
plt.xlabel('0 = False(Embarazada No fumadora)   1 = True(Embarazada Fumadora)')
plt.ylabel('Número de casos')
plt.title('Cantidad de embarazadas fumadoras y no fumadoras escogidas al azar')
plt.show()
```

```python
#PREGUNTA 5 LETRA B): A PARTIR DEL VALOR ANTERIOR, CONSTRUYA UN MODELO DE DISTRIBUCIÓN BINOMINAL QUE MODELE EL EXPERIMENTO
#CORRESPONDIENTE A ESCOGER 5 MADRES AL AZAR Y OBSERVAR CUÁNTAS FUMAN.

#1ª PARTE: COSTRUCCIÓN DE MODELO BINOMIAL.
fumadora = df_baby['Maternal.Smoker'].dropna()
fuma = np.where(fumadora, 1, 0)
p = sum(fuma) / len(fuma)

binom = []
binom.append(np.mean(np.random.binomial(5, p, size = 1)))

#2ª PARTE: GRAFICACIÓN DE LOS RESULTADOS.
print(binom)
plt.hist(x=binom, alpha=0.5, color='blue')
plt.title('Cantidad de embarazadas fumadoras escogidas al azar')
plt.xlabel('Selección de embrazadas')
plt.ylabel('Número de casos de embarazadas fumadoras')
```

```python
#PREGUNTA 5 LETRA C): UTILIZANDO ESTE MODELO, SIMULA 1000 MUESTRAS DE TAMAÑO 8, Y CALCULA SUS MEDIAS MUESTRALES.


#PASO 1: Código para calcular la probabilidad de embarazadas fumadoras.
df_mothers_smoker = df_baby.loc[df_baby['Maternal.Smoker'] == True]['Maternal.Smoker'].sum()
casos = df_baby['Maternal.Smoker'].size
p = df_mothers_smoker / casos

p = p /100                                          #Porcentaje de embarazadas fumadoras dividido por el total de embarazadas.
n = 8                                               #Número de casos.

medias_muestras_simuladas = []
for _ in range(1000):

    muestra = np.random.binomial(n, p, size=1)      #Simulaciòn de 1000 muestras tamaño 8.
    media_muestra = np.mean(muestra)                #Cálculo de la edia de las muestras.
    medias_muestras_simuladas.append(media_muestra)           #Agregación de elementos a la lista mediante función append.


#PASO 2: graficación de los resultados.
plt.hist(medias_muestras_simuladas, color='red', alpha=0.5)
plt.xlabel('Media muestral')
plt.ylabel('Frecuencia')
plt.title('Modelo de simulación de 1000 medias muestrales')
plt.show()

#PASO 3: Cálculo de la media y de la desviación estándar de las medias muestrales calculadas previamente.
media_medias_muestras = np.mean(medias_muestras)
desviacion_estandar_medias_muestras = np.std(medias_muestras)

print(f'Media de las muestras: {media_medias_muestras:.2f}')
print(f'Desviación estándar de las medias muestrales: {desviacion_estandar_medias_muestras:.2f}')

```
